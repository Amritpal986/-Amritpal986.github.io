```python
%%js
function predictGrowth(revenues){

    let growth = 0;

    for(let i = 1; i < revenues.length; i++){
        growth += revenues[i] - revenues[i-1];
    }

    growth = growth / (revenues.length - 1);

    return Math.round(revenues[revenues.length-1] + growth);
}
```
